## Prepare AI-ready data

```python
df = spark.read.table("ecommerce_catalog.bronze.events_nov")
df.printSchema()
```

```python
from pyspark.sql.functions import col

df_clean = (
    df.withColumnRenamed("event_time", "event_timestamp")
      .withColumnRenamed("category_code", "category")
)
```

```python
df_clean.write.format("delta") \
    .mode("overwrite") \
    .saveAsTable("ecommerce_catalog.silver.events_clean")
```

```python
%sql
select * from ecommerce_catalog.silver.events_clean
```
