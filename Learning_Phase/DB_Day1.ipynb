```python
# Create a simple DataFrame
data = [("Iphone", 9000), ("Galaxy", 800), ("Pixel", 7000)]

df = spark.createDataFrame(data, ["product", "price"])
df.show()
```

```python
df.filter(df.price>1000).show()
```
